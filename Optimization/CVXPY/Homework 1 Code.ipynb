```python
import cvxpy as cp
```

```python
#Problem1

x = cp.Variable(3, nonneg = True)

obj_func = 600*x[0] + 460*x[1] + 320*x[2]

constraints = []
constraints.append(x[0]<=600000) #Demand Constraint
constraints.append(x[1]<=800000) #Demand Constraint
constraints.append(x[2]<=700000) #Demand Constraint
constraints.append(18*x[0]+29*x[1]+38*x[2]>=10) #Government mpg regulation

problem = cp.Problem(cp.Maximize(obj_func), constraints)

problem.solve(solver=cp.GUROBI, verbose = True)

print("obj_func = ")
print(obj_func.value)
print("x = ")
print(x.value)
```

```python
#Problem 3

O = cp.Variable(6, nonneg=True) #One year investment
T = cp.Variable(4, nonneg=True) #Three year investment
F = cp.Variable(2, nonneg=True) #Five year investment

obj_func = 1.051*O[5] + 1.162*T[3] + 1.285*F[1]

constraints = []
constraints.append(1000 == O[0] + T[0] + F[0]) #Year 1
constraints.append(1.051*O[0] == O[1] + T[1] + F[1]) #Year 2
constraints.append(1.051*O[1] == O[2] + T[2]) #Year 3
constraints.append(1.051*O[2] + 1.162*T[0] == O[3] + T[3]) #Year 4
constraints.append(1.051*O[3] + 1.162*T[1] == O[4]) #Year 5
constraints.append(1.051*O[4] + 1.162*T[2] + 1.285*F[0]== O[5]) #Year 5

problem = cp.Problem(cp.Maximize(obj_func), constraints)

problem.solve(solver=cp.GUROBI, verbose = True)

print("obj_func = ")
print("$", obj_func.value)
print("One Year Investment = ")
print(O.value)
print("Three Year Investment = ")
print(T.value)
print("Five Year Investment = ")
print(F.value)
```

```python
#Problem 5

W = cp.Variable(3, nonneg=True) #One year investment
G = cp.Variable(4, nonneg=True) #Three year investment

obj_func = 1.35*W[0] + 1.28*W[1] + 1.47*W[2] + 1.14*G[0] + 1.19*G[1] + 1.26*G[2] + 1.16*G[3]

constraints = []
constraints.append(W[0]+W[1]+W[2] == 320) #Widget Demand Constraint
constraints.append(G[0]+G[1]+G[2]+G[3] == 250) #Garget Demand Constraint
constraints.append(0.25*W[0]+0.25*W[2]+0.26*G[1]+0.20*G[2] <= 80) #Center 1 Time Constraint
constraints.append(0.13*W[0]+0.34*W[1]+0.18*G[0]+0.12*G[3] <= 80) #Center 2 Time Constraint
constraints.append(0.15*W[1]+0.42*W[2]+0.22*G[1]+0.18*G[3] <= 80) #Center 3 Time Constraint
constraints.append(0.20*W[0]+0.30*G[2]+0.20*G[3] <= 80) #Center 4 Time Constraint
constraints.append(0.28*W[1]+0.32*G[0]+0.14*G[1] <= 80) #Center 5 Time Constraint


problem = cp.Problem(cp.Minimize(obj_func), constraints)

problem.solve(solver=cp.GUROBI, verbose = True)

print("obj_func = ")
print("$", obj_func.value)
print("Widgets = ")
print(W.value)
print("Gargets = ")
print(G.value)
```
