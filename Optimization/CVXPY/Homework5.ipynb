```python
import cvxpy as cp
```

```python
#Problem1
#c

# Define the variables
x1 = cp.Variable(nonneg=True)
x2 = cp.Variable(nonneg=True)

# Define the objective function (minimizing -16x1 - 15x2)
objective = cp.Minimize(-16 * x1 - 15 * x2)

# Define the constraints
constraints = [
    40 * x1 + 31 * x2 <= 124,
    -x1 + x2 <= 1,
    x1 <= 3
]

# Set up the problem and solve it
problem = cp.Problem(objective, constraints)
problem.solve(solver=cp.GUROBI, verbose = True)

# Display the results
print("Optimal value of z:", problem.value)
print("Values of x1 and x2:", x1.value, x2.value)
```

```python
#Problem3
#b

import cvxpy as cp

# Define the variables
x1 = cp.Variable(nonneg=True)
x2 = cp.Variable(nonneg=True)
x3 = cp.Variable(nonneg=True)

# Define the objective function: Minimize −2x1 − 2x2 − 4x3
objective = cp.Minimize(-2 * x1 - 2 * x2 - 4 * x3)

# Define the constraints
constraints = [
    2 * x1 + x2 + x3 <= 2,   # First constraint
    3 * x1 + 4 * x2 + 2 * x3 >= 8  # Second constraint (rewritten for standard form)
]

# Set up the problem and solve it
problem = cp.Problem(objective, constraints)
problem.solve(solver=cp.GUROBI, verbose = True)

# Display the results
print("Optimal value of z:", problem.value)
print("Values of x1, x2, and x3:", x1.value, x2.value, x3.value)
```

```python
# Define integer variables
x1 = cp.Variable(integer=True)
x2 = cp.Variable(integer=True)
x3 = cp.Variable(integer=True)

# Define the objective function: Minimize −2x1 − 2x2 − 4x3
objective = cp.Minimize(-2 * x1 - 2 * x2 - 4 * x3)

# Define the constraints
constraints = [
    2 * x1 + x2 + x3 <= 2,        # Constraint 1
    3 * x1 + 4 * x2 + 2 * x3 >= 8,  # Constraint 2
    x1 >= 0,                       # Non-negativity for x1
    x2 >= 0,                       # Non-negativity for x2
    x3 >= 0                        # Non-negativity for x3
]

# Set up the problem
problem = cp.Problem(objective, constraints)

# Solve the problem using a solver that supports integer constraints
problem.solve(solver=cp.GLPK_MI)  # or another MIP-compatible solver

# Display the results
print("Optimal value of z:", problem.value)
print("Values of x1, x2, and x3:", x1.value, x2.value, x3.value)
```
